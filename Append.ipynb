```python
class Student():
    def __init__(self,name,age):
        self.name=name
        self.age=age

        #function
    def school(self):
        print(f'my name is {self.name}')
        
```

```python
# Multiple inheritance
#parent
class father():
    def house(self):
        print("Father has a house")
# Parent         
class mother():
    def land(self):
        print("Mother has land")
# Child
class son(father,mother):
    def bike(self):
        print("son has a bike")
```

```python
s=son()
s.house()
s.land()
s.bike()
```

```python
#single inheritance
# parent
class animal():
    def printer(self):
        print("I am an animal")
# child
class pet():
    def bark(self):
        print("Dog barks")
```

```python
A=animal()
A.printer()
A=pet()
A.bark()
```

```python
class Grandfather():
    def land(self):
        print("Grandfather is sick")
class father(Grandfather):
    def house(self):
        print("father has a car")
class son(father):
    def bike(self):
        print("son is vey sad")                
```

```python
h=son()
h.land()
h.house()
h.bike()
```

```python
class Animal():
    def Stay(self):
        print("I am an animal")
class Dog(Animal):
    def Bark(self):
        print("Dog guards the house")
class Cat(Animal):
    def meow(self):
        print("Cat is a pet")                
```

```python
D=Dog()
D.Stay()
D.Bark()
```

```python
C=Cat()
C.Stay()
C.meow()
```

```python
#Hybrid
class A():
    def methodA (self):
        print("class A")
class B(A):
    def methodB(self): 
        print("class B")       
class C(A):
    def methodC(self): 
        print("class C")       
class D(B,C):
    def methodD(self): 
        print("class D")       
            
```

```python
obj=D()
obj.methodD()
```
